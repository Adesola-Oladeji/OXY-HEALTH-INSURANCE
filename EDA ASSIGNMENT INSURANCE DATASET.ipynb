```python
# Group bmi
def bmi_group(x):
    if x < 18.5:
        return "<18.5 Underweight"
    elif x <= 24.9:
        return ">=18.5 - 24.9 healthy"
    elif x <= 29.9:
        return "25 - 29.9 overweight"
    elif x <= 39.9:
        return "30 - 39.9 obesity"
    else:
        return ">=40 severe obesity"
    
 # Apply function to the data
df["bmi_group"] = df["bmi"].apply(bmi_group)
df.head()
```

```python
### Total transactions by customer Bmi Group

plt.figure(figsize=(15, 5))

ax = sns.countplot(y=df["bmi_group"], order=df["bmi_group"].value_counts(ascending=False).index)

# Get values for each bar to label
values = df["bmi_group"].value_counts(ascending=False).values

# Add labels on top of the bars
ax.bar_label(ax.containers[0], labels=values, fontsize=12, color='black', label_type='edge')

plt.show()
```

### OBSERVATIONS:
It was observed that 623 customers were Obesity, 380 were overweght, 222 were healthy. 
93 customers were having severe obesity while 20 were underweight

```python
### Total transactions by customer Age Group
plt.figure(figsize=(15, 5))

ax = sns.countplot(y=df["age_group"], order=df["age_group"].value_counts(ascending=False).index)
values = df["age_group"].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=values);
```

```python
# BIVARIATE ANALYSIS
Bivariate analysis involves analyzing the relationship between two variables
. focus on Charges
```

```python
df.columns
```

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(20,10))
age_charge = df.groupby("age_group")["Charges"].sum().reset_index()
sns.barplot(x='age_group', data=age_charge, y='Charges', ax=axs[0,0])
axs[0,0].set_title("age_group by Charges")

bmi_charge = df.groupby("bmi_group")["Charges"].sum().reset_index()
sns.barplot(x='bmi_group', data=bmi_charge, y='Charges', ax=axs[0,1])
axs[0,1].set_title("bmi_group by Charges")

reg_charge = df.groupby("Region")["Charges"].sum().reset_index()
sns.barplot(x='Region', data=reg_charge, y='Charges', ax=axs[0,2])
axs[0,2].set_title("Region by Charges")

smok_charge = df.groupby("Smoker")["Charges"].sum().reset_index()
sns.barplot(x='Smoker', data=smok_charge, y='Charges', ax=axs[1,0])
axs[0,0].set_title("Smoker by Charges")

Child_charge = df.groupby("children")["Charges"].sum().reset_index()
sns.barplot(x='children', data=Child_charge, y='Charges', ax=axs[1,1])
axs[0,0].set_title("children by Charges")

sex_charge = df.groupby("Sex")["Charges"].sum().reset_index()
sns.barplot(x='Sex', data=sex_charge, y='Charges', ax=axs[1,2])
axs[0,0].set_title("Sex by Charges");
```

#### OBSERVATIONS
1. Elders paid the highest charges while the young adult paid the least charges
2. The obesed customers paid the higest charges, followed by the overweight customers 
    while the underweight customers were charged the lowest.
3. The region that were charged most were Southeast while Northwest paid the least charges.
4. There is no significant difference in the charges of yes and no smokers
5. Customers with no children were charged more and the customers with 5 children got the least charges
6. Male customers paid more charges than the female customers.
